## MODULO MOEDA

```python
def aumentar(preco, taxa):
    res = preco + (preco * taxa / 100)
    return f'R${res:.2f}'.replace('.',',')

def diminuir(preco, taxa):
    res = preco - (preco * taxa / 100)
    return f'R${res:.2f}'.replace('.',',')

def dobro(preco):
    res = preco * 2
    return f'R${res:.2f}'.replace('.',',') 

def metade(preco):
    res = preco / 2
    return f'R${res:.2f}'.replace('.',',')
```
